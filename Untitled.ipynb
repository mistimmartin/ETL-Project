```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
# from config import password
```

```python
calendar_df = pd.read_csv("Resources/calendar.csv")
calendar_df = calendar_df[calendar_df.available != 'f']
calendar_df
```

```python
listings_df = pd.read_csv("Resources/listings.csv")
listings_df.head()
```

```python
reviews_df = pd.read_csv("Resources/reviews.csv")
reviews_df.head()
```

```python
calendar_df_clean= calendar_df[["listing_id","date","price"]].copy()
calendar_df_clean.count()
```

```python
listings_df_clean=listings_df[["id","name","neighbourhood_cleansed","price","number_of_reviews","review_scores_rating"]].copy()
listings_df_clean.head()
```

```python
reviews_df_clean= reviews_df[["listing_id","comments"]].copy()
reviews_df_clean.head()
```

```python
connection_string = "postgres:Sunshine1@localhost:5432/seattle_airbnb"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
calendar_df_clean.to_sql(name='calendar', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from calendar', con=engine).head()
```

```python
listings_df_clean.to_sql(name='listings',con=engine,if_exists='append',index=False)
```

```python
pd.read_sql_query('select * from listings',con=engine).head()
```

```python
reviews_df_clean.to_sql(name='reviews',con=engine,if_exists='append',index=False)
```

```python
pd.read_sql_query('select * from reviews',con=engine).head()
```

```python
combined_df = pd.read_sql_query('select c.listing_id, c.price, l.name, l.number_of_reviews, l.review_scores_rating, r.comments from calendar as c\
                                join listings as l\
                                on c.listing_id = l.id\
                                join reviews as r\
                                on c.listing_id = r.listing_id', con=engine)
```
